# Questão 3

```python
def genRing(n):
    return array([[(i-1) % n, (i+1) % n] for i in range(n)])

def genComplete(n):
    return array([array(range(n))[arange(n) != i] for i in range(n)])

def genTree(levels):
    n = 2**levels - 1
    def genEdges(i):
        aux = array([floor((i-1)/2), 2*i+1, 2*i+2], dtype=int)
        return aux[(aux >= 0) & (aux < n)]
    return array([genEdges(i) for i in range(n)])  

def genGrid(a, b):
    def genEdges(i, j):
        n = lambda k, l: k * b + l
        aux = []
        if i > 0:
            aux.append(n(i-1, j))
        if j > 0:
            aux.append(n(i, j-1))
        if i < a-1:
            aux.append(n(i+1, j))
        if j < b-1:
            aux.append(n(i, j+1))
        return aux
    return array([genEdges(i, j) for i in range(a) for j in range(b)])
```

```python
def πTree(Adj):
    return array([len(i) for i in Adj]) / (2 * (len(Adj) - 1))

def πGrid(Adj, a, b):
    return array([len(i) for i in Adj]) / (2 * a * b - a - b) / 2
```

```python
def adj2matrix(Adj):
    n = len(Adj)
    M = zeros((n, n))
    for v, edges in enumerate(Adj):
        for u in edges:
            M[v, u] = 1
    return M
```

```python
Adjs = {
    'Anel': genRing(100),
    'Árvore': genTree(7),
    'Grid 10x10': genGrid(10, 10),
    'Grid 100x1': genGrid(100, 1),
    'Completo': genComplete(100),
}

graphs = {i: adj2matrix(Adjs[i]) for i in Adjs}

ts = 100000
πₜ = {i: zeros((ts, len(graphs[i]))) for i in graphs}

for i in graphs:
    graphs[i] += diag(graphs[i].sum(1))
    graphs[i] = (graphs[i].T / graphs[i].sum(1)).T
    
    # eu optei por fazer a multiplicação de matriz direta
    # pois precisamos de todos os πₜ e desta maneira
    # temos menos erro numérico e melhor desempenho
    πₜ[i][0] = graphs[i][0]
    for j in range(1, ts):
         πₜ[i][j] = πₜ[i][j-1] @ graphs[i]

    # codigo fazendo por decomposição espectral
#     l, Q = eig(graphs[i])
#     Qᵢ = inv(Q)
#     for j in range(ts):
#         πₜ[i][j] = (Q @ diag(l ** j) @ Qᵢ)[0]
```

```python
realπ = {
    'Anel': 1/100,
    'Árvore': πTree(Adjs["Árvore"]),
    'Grid 10x10': πGrid(Adjs["Grid 10x10"], 10, 10),
    'Grid 100x1': πGrid(Adjs["Grid 100x1"], 100, 1),
    'Completo': 1/100,
}
```

```python
figure(figsize=(16,8))

xlabel('$t$')
ylabel('$d_{TV}(\pi_t, \pi)$')

for i in graphs:
    loglog([dtv(j, realπ[i]) for j in πₜ[i]], label=i)

legend();
```

# Questão 4

```python
def findτ(M, ϵ, πₒ, realπ, tₒ=0, step=100, tol=1e-12):
    # a ideia aqui é encontrar τ_ϵ pelo método da bisseção
    l, Q = eig(M)
    Qᵢ = inv(Q)
    Qaux = πₒ @ Q
    
    # encontra o intervalo on τ_ϵ se encontra
    a = tₒ
    b = tₒ + step
    δ = dtv(Qaux @ diag(l ** b) @ Qᵢ, realπ)
    while δ > ϵ:
        a += step
        b += step
        δ = dtv(Qaux @ diag(l ** b) @ Qᵢ, realπ)

    while b - a > 1:
        m = int((a + b) / 2)
        δ = dtv(Qaux @ diag(l ** m) @ Qᵢ, realπ)
        if abs(δ - ϵ) < tol:
            break
        if δ > ϵ:
            a = m
        else:
            b = m
    if b - a > 1:
        return m
    return b

def findτopt(l, Q, Qᵢ, ϵ, k, realπ, tₒ=0, step=100, tol=1e-12):
    # a ideia aqui é encontrar τ_ϵ pelo método da bisseção
    Qaux = Q[k]
    
    # encontra o intervalo on τ_ϵ se encontra
    a = tₒ
    b = tₒ + step
    δ = dtv((Qaux * (l ** b)) @ Qᵢ, realπ)
    while δ > ϵ:
        a += step
        b += step
        δ = dtv((Qaux * (l ** b)) @ Qᵢ, realπ)

    while b - a > 1:
        m = int((a + b) / 2)
        δ = dtv((Qaux * (l ** b)) @ Qᵢ, realπ)
        if abs(δ - ϵ) < tol:
            break
        if δ > ϵ:
            a = m
        else:
            b = m
    if b - a > 1:
        return m
    return b
```

```python
ϵ = 1e-4
Ns = array([10, 50, 100, 300, 700, 1000])
grids = array([[5,2], [5, 10], [10, 10], [15, 20], [35, 20], [40, 25]])
```

```python
Adjs = {
    'Anéis': [genRing(n) for n in Ns],
    'Árvores': [genTree(int(ceil(log2(n)))) for n in Ns],
    'Grids': [genGrid(*i) for i in grids],
}

Ms = {i: [adj2matrix(j) for j in Adjs[i]] for i in Adjs}

for i in Ms:
    for j, k in enumerate(Ms[i]):
        Ms[i][j] += diag(k.sum(1))
        Ms[i][j] = (k.T / k.sum(1)).T

realπ = {
    'Anéis': 1 / Ns,
    'Árvores': [πTree(j) for j in Adjs["Árvores"]],
    'Grids': [πGrid(j, *i) for i,j in zip(grids, Adjs["Grids"])],
}
```

```python
τs = []

for i, a in enumerate(Adjs):
    print(a)
    aux = []
    for j, n in enumerate(Ns):
        l, Q = eig(Ms[a][j])
        Qᵢ = inv(Q)
        print(n, end='')
        now = time()
        aux.append([findτopt(l, Q, Qᵢ, ϵ, k, realπ[a][j], step=1000000) for k in range(len(Ms[a][j]))])
        print(" - {:.6f}".format(time()-now))
    τs.append(aux)
```

```python
figure(figsize=(16, 8))

for i, j in enumerate(Adjs):
    loglog([len(k) for k in Ms[j]], [max(k) for k in τs[i]], label=j);

xlabel('$n$')
ylabel(r'$\tau_n$')
legend();
```
